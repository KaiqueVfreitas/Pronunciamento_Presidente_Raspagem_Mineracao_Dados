```python
class Pronunciamento:
    def __init__(self, bs_artigo=None):
        self.url = bs_artigo.find(class_='summary url')["href"]
```

```python
def pronuncamento_organizado_por_data(datestr):
    # Precisa ser retornada invertida para comportar no banco de dados
    return "-".join(reversed(datestr.split("/")))
```
